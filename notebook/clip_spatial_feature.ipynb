```python
import clip
import os
import torch
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
clip_model, clip_preprocess = clip.load("ViT-B/32", device='cuda')
```

```python
original_images = []
images = []
for pic_name in os.listdir('images/'):
#     print(pic_name)
    if pic_name.endswith('.jpg'):
        image = Image.open(os.path.join('images/', pic_name))
        original_images.append(image)
        images.append(clip_preprocess(image))
```

```python
plt.figure(figsize=(10,8))
for num_img in range(9):
    plt.subplot(3,3,num_img+1)
    plt.imshow(original_images[num_img])
```

```python
descriptions = [
    [ # 1
        'a photo of a cat.',
        'a photo of a shelf.',
        'a photo where a cat lying on a shelf.',
        'a photo where a cat lying under a shelf',
        'a photo where a cat lying on the left of a shelf'
    ],
    [ # 2
        'a photo of a cat.',
        'a photo of a table.',
        'a photo where a cat lying on or above a table.',
        'a photo where a cat lying under a table.',
        'a photo where a cat lying on the left of a table.'
    ],
    [ # 3
        'a photo of an apple.',
        'a photo of two apples.',
        'a photo where the red apple is on the left of the green apple.',
        'a photo where the red apple is on the right of the green apple.',
        'a photo where the green apple is on the right of the red apple.'
    ],
    [ # 4
        'a photo of a piano.',
        'a photo of two cats.',
        'a photo of two cats, the darker one lying on a bench.',
        'a photo of two cats, the darker one lying under a bench.',
        'a photo of two cats, both lying under a bench.',
    ],
    [ # 5
        'a photo of a table.',
        'a photo of a child.',
        'a photo of a child hidding under a table.',
        'a photo of a child standing on a table.',
        'a photo of a child and a table.'
    ],
    [ # 6
        'a photo of two animals.',
        'a photo of a dog and a goat.',
        'a photo of two animals, the dog standing on the right of the goat.',
        'a photo of two animals, the dog standing on the left of the goat.',
        'a photo where the dog is on the right of the goat.'
    ],
    [ # 7
        'a photo of a bench and a table.',
        'a bench under the white table.',
        'a bench on the table.',
        'a photo of some furnitures.',
        'a bench on the left of a table.'
    ],
    [ # 8
        'a photo of two animals.',
        'a photo of two rabbits.',
        'a white rabbit on the right of a brown rabbit.',
        'a white rabbit on the left of a brown rabbit.',
        'a white rabbit on a brown rabbit.',
    ],
    [ # 9
        'a photo of some animals.',
        'a photo of a rabbit and two ducks.',
        'a rabbit on the right of two ducks.',
        'a rabbit on the left of two ducks.',
        'a photo of two ducks on the left of a rabbit.',
    ],
]

plt.figure(figsize=(14, 50))
for i, descs in enumerate(descriptions):
    plt.subplot(9,1,i+1)
    
    image_input = images[i].unsqueeze(0).cuda()
    text_tokens = clip.tokenize(["This is " + desc for desc in descs]).cuda()
    
    with torch.no_grad():
        image_features = clip_model.encode_image(image_input).float()
        text_features = clip_model.encode_text(text_tokens).float()
#     print(image_features.shape, text_features.shape)
    
    image_features /= image_features.norm(dim=-1, keepdim=True)
    text_features /= text_features.norm(dim=-1, keepdim=True)
    similarity = text_features.cpu() @ image_features.cpu().T
#     print(similarity.shape)
    
    count = len(descs)
#     plt.figure(figsize=(4, 8))
    plt.imshow(similarity, vmin=0.2, vmax=0.35)
    
    # plt.colorbar()
    plt.yticks(range(count), descs, fontsize=12)
    plt.xticks([])
#     for i, image in enumerate(original_images):
    plt.imshow(original_images[i], extent=(-0.5, 0.5, -1.6, -0.6), origin="lower")

        
    for x in range(similarity.shape[1]):
        for y in range(similarity.shape[0]):
            plt.text(x, y, f"{similarity[y, x]:.2f}", ha="center", va="center", size=8)

    for side in ["left", "top", "right", "bottom"]:
        plt.gca().spines[side].set_visible(False)

    plt.xlim([-0.5, 0.5])
    plt.ylim([count, -2])

    plt.title("Cosine similarity between text and image features", size=12)
#     assert False
```
