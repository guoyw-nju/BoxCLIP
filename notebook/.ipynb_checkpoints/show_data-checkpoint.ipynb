```python
# coco = CocoDetection(root=root, annFile=annFile, annFile_cap=annFile_cap, transform=transforms.ToTensor())
coco_without_transform = CocoDetection(root=root, annFile=annFile, annFile_cap=annFile_cap)
coco = CocoDetection(root=root, annFile=annFile, annFile_cap=annFile_cap, 
                     transform=transforms.Compose([transforms.Resize([224, 224]), clip_preprocess]))
```

```python
to_pil_image = transforms.ToPILImage()

img_num = 30
plt.figure(figsize=(16, 6))

plt.subplot(2,2,1)
plt.imshow(coco_without_transform[img_num][0])
plt.title('original img', color='white')

    
plt.subplot(2,2,2)
plt.imshow(clip_preprocess(coco_without_transform[img_num][0]).permute(1,2,0))
plt.title('clip_pre only', color='white')
for obj in coco_without_transform[img_num][1]:
    [bbox_x, bbox_y, bbox_w, bbox_h] = obj['bbox']
    plt.gca().add_patch(plt.Rectangle((bbox_x*224, bbox_y*224), bbox_w*224, bbox_h*224, 
                                      fill=False, edgecolor='r', linewidth=2))
    
plt.subplot(2,2,3)
plt.imshow(coco[img_num][0].permute(1,2,0))
plt.title('clip_pre with resize', color='white')
for obj in coco[img_num][1]:
    [bbox_x, bbox_y, bbox_w, bbox_h] = obj['bbox']
    plt.gca().add_patch(plt.Rectangle((bbox_x*224, bbox_y*224), bbox_w*224, bbox_h*224, 
                                      fill=False, edgecolor='r', linewidth=2))
```

```python
data_loader = torch.utils.data.DataLoader(coco, batch_size=2, shuffle=True, collate_fn=collate_fn_coco)

num_image = 2
plt.figure(figsize=(16,10))
for batch in data_loader:    
    for i in range(num_image):
        plt.subplot(2,2,i+1)
        plt.imshow(batch['clip_images'][i].permute(1,2,0).cpu())
        
        for j in range(batch['bboxs'].shape[1]):
            [bbox_x, bbox_y, bbox_w, bbox_h] = batch['bboxs'][i][j]
            plt.gca().add_patch(plt.Rectangle((bbox_x*224, bbox_y*224), bbox_w*224, bbox_h*224, 
                                              fill=False, edgecolor='r', linewidth=2))
        
        print(batch['cat_name'][i])
        plt.title('\n'.join(batch['clip_texts'][i]), color='white')
        
    break
```
